```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.array([1, 4, 8, 10]) # array for x
y_1 = np.array([0.0000005226, 0.004028, 0.9028, 1.698]) # array for y

x_2 = np.array([1, 4, 8, 10]) # array for x
y_2 = np.array([0.0000005243, 0.004984, 0.9328, 1.732]) # array for y

x_3 = np.array([1, 4, 8, 10]) # array for x
y_3 = np.array([0.0000004996, 0.00004973, 0.2938, 0.8986]) # array for y

x_4 = np.array([1, 4, 8, 10]) # array for x
y_4 = np.array([0.0000005412, 0.00005428, 0.3467, 0.9796]) # array for y

plt.figure()
```

```python
plt.plot(x_1, y_1,marker = 'o',label = "Ratio A")  #5-5-90%
plt.plot(x_2, y_2,marker = '+',label = "Ratio B") #5-15-80%
plt.plot(x_3, y_3,marker = 'x',label = "Ratio C") #10-20-70%
plt.plot(x_4, y_4,marker = 's',label = "Ratio D") #10-40-50%

plt.xlabel('Electron Energy (MeV)')
plt.ylabel('Particle Flux (cm\N{SUPERSCRIPT TWO}/source)')
plt.legend()
plt.show()
```
