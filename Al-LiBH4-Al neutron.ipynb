```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.array([10, 20, 400, 500]) # array for x
y_1 = np.array([0.0001398, 0.005988, 0.1636, 0.164]) # array for y

x_2 = np.array([10, 20, 400, 500]) # array for x
y_2 = np.array([0.0001354, 0.006139, 0.138, 0.164]) # array for y

x_3 = np.array([10, 20, 400, 500]) # array for x
y_3 = np.array([0.0001379, 0.005999, 0.132, 0.156]) # array for y

x_4 = np.array([10, 20, 400, 500]) # array for x
y_4 = np.array([0.0001346, 0.005993, 0.0781, 0.0953]) # array for y

x_5 = np.array([10, 20, 400, 500]) # array for x
y_5 = np.array([0, 0, 0, 0]) # array for y

x_6 = np.array([10, 20, 400, 500]) # array for x
y_6 = np.array([0.0001462, 0.005726, 0.1931, 0.228]) # array for y

plt.figure()
```

```python
plt.plot(x_5, y_5,marker = '*',label = "Al") #blue Al
plt.plot(x_6, y_6,marker = 'D',label = "No Shield") #blue Al
plt.plot(x_1, y_1,marker = 'o',label = "SWC-A")  #1-9-1%
plt.plot(x_2, y_2,marker = '+',label = "SWC-B") #3-6-1%
plt.plot(x_3, y_3,marker = 'x',label = "SWC-C") #1-3-6%
plt.plot(x_4, y_4,marker = 's',label = "SWC-D") #4-2-4%

plt.xlabel('Proton Energy (MeV)')
plt.ylabel('Particle Flux (cm\N{SUPERSCRIPT TWO}/source)')

plt.legend()
plt.show()
```
