```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.array([1, 4, 8, 10]) # array for x
y_1 = np.array([0.000234, 0.008699, 0.04025, 0.0588]) # array for y

x_2 = np.array([1, 4, 8, 10]) # array for x
y_2 = np.array([0.0001562, 0.008347, 0.0362, 0.0516]) # array for y

x_3 = np.array([1, 4, 8, 10]) # array for x
y_3 = np.array([0.0002453, 0.008867, 0.0363, 0.0514]) # array for y

x_4 = np.array([1, 4, 8, 10]) # array for x
y_4 = np.array([0.0001462, 0.006528, 0.0211, 0.028]) # array for y

x_5 = np.array([1, 4, 8, 10]) # array for x
y_5 = np.array([0.0002812, 0.008783, 0.04262, 0.0642]) # array for y

x_6 = np.array([1, 4, 8, 10]) # array for x
y_6 = np.array([0, 0, 0, 0]) # array for y

plt.figure()
```

```python
plt.plot(x_5, y_5,marker = '*',label = "Al") #blue Al
plt.plot(x_6, y_6,marker = 'D',label = "No Shield") #blue Al
plt.plot(x_1, y_1,marker = 'o',label = "Ratio A")  #1-9-1%
plt.plot(x_2, y_2,marker = '+',label = "Ratio B") #3-6-1%
plt.plot(x_3, y_3,marker = 'x',label = "Ratio C") #1-3-6%
plt.plot(x_4, y_4,marker = 's',label = "Ratio D") #4-2-4%

plt.xlabel('Photon Energy (MeV)')
plt.ylabel('Particle Flux (cm\N{SUPERSCRIPT TWO}/source)')
plt.legend()
plt.show()
```
