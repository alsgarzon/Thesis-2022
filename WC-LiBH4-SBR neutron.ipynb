```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.array([10, 20, 400, 500]) # array for x
y_1 = np.array([0.0000814, 0.00327, 0.1408, 0.1684]) # array for y

x_2 = np.array([10, 20, 400, 500]) # array for x
y_2 = np.array([0.00008286, 0.003312, 0.1386, 0.165]) # array for y

x_3 = np.array([10, 20, 400, 500]) # array for x
y_3 = np.array([0.00008622, 0.003474, 0.2011, 0.2399]) # array for y

x_4 = np.array([10, 20, 400, 500]) # array for x
y_4 = np.array([0.0000846, 0.003502, 0.1973, 0.2343]) # array for y

x_5 = np.array([10, 20, 400, 500]) # array for x
y_5 = np.array([0.0001421, 0.005928, 0.1931, 0.2273]) # array for y

plt.figure()
```

```python
plt.plot(x_5, y_5,marker = '*',label = "Al") #blue Al
plt.plot(x_1, y_1,marker = 'o',label = "Ratio A")  #5-5-90%
plt.plot(x_2, y_2,marker = '+',label = "Ratio B") #5-15-80%
plt.plot(x_3, y_3,marker = 'x',label = "Ratio C") #10-20-70%
plt.plot(x_4, y_4,marker = 's',label = "Ratio D") #10-40-50%

plt.xlabel('Neutron Energy (MeV)')
plt.ylabel('Particle Flux (cm\N{SUPERSCRIPT TWO}/source)')


plt.legend()
plt.show()
```
