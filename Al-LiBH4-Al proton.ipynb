```python
x_1 = np.array([10, 20, 400, 500]) # array for x
y_1 = np.array([0, 0.0000003358, 161.4, 161.4]) # array for y

x_2 = np.array([10, 20, 400, 500]) # array for x
y_2 = np.array([0, 0.00000129, 161.5, 145.6]) # array for y

x_3 = np.array([10, 20, 400, 500]) # array for x
y_3 = np.array([0, 0.0000008055, 161.5, 145.5]) # array for y

x_4 = np.array([10, 20, 400, 500]) # array for x
y_4 = np.array([0, 0.000002642, 161.8, 145.8]) # array for y

x_5 = np.array([10, 20, 400, 500]) # array for x
y_5 = np.array([0, 0, 161.2, 145.3]) # array for y

x_6 = np.array([10, 20, 400, 500]) # array for x
y_6 = np.array([18.6, 37.2, 162, 146]) # array for y

plt.figure()
```
