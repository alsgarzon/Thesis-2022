```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.array([10, 20, 400, 500]) # array for x
y_1 = np.array([0, 0.0000003203, 161.7, 145.7]) # array for y

x_2 = np.array([10, 20, 400, 500]) # array for x
y_2 = np.array([0, 0.0000003234, 161.7, 145.7]) # array for y

x_3 = np.array([10, 20, 400, 500]) # array for x
y_3 = np.array([0, 0.0000002643, 161.5, 145.5]) # array for y

x_4 = np.array([10, 20, 400, 500]) # array for x
y_4 = np.array([0, 0.0000001488, 161.5, 145.5]) # array for y

x_5 = np.array([10, 20, 400, 500]) # array for x
y_5 = np.array([0, 0.0000001092, 161.2, 145.3]) # array for y

plt.figure()
```
